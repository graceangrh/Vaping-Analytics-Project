```python
import pandas as pd
import numpy as np
import nltk
```

```python
int_data = pd.read_excel('SA_cleaned_interview_vaper_friend.xlsx', header=0)
```

```python
print(int_data.head())
print("\nNumber of records in train  : ", len(int_data))
```

Extract the Text column from the data frame.

```python
better_alt = int_data["Is vaping a better alternative to smoking?"]
print (better_alt)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
print (analyzer)
```

```python
#nltk.download('vader_lexicon')
```

```python
[print(i, analyzer.polarity_scores(better_alt.iloc[i])) for i in range(len(better_alt))]
```

```python
my_vader_score_compound = [ ] 
my_vader_score_positive = [ ] 
my_vader_score_negative = [ ] 
my_vader_score_neutral = [ ] 

for i in range(len(better_alt)):
    my_analyzer = analyzer.polarity_scores(better_alt.iloc[i])
    my_vader_score_compound.append(my_analyzer['compound'])
    my_vader_score_positive.append(my_analyzer['pos'])
    my_vader_score_negative.append(my_analyzer['neg']) 
    my_vader_score_neutral.append(my_analyzer['neu']) 

my_vader_score_compound = np.array(my_vader_score_compound)
my_vader_score_positive = np.array(my_vader_score_positive)
my_vader_score_negative = np.array(my_vader_score_negative)
my_vader_score_neutral = np.array(my_vader_score_neutral)
```

```python
int_data['compound'] = my_vader_score_compound
int_data['postve'] = my_vader_score_positive
int_data['neg'] = my_vader_score_negative
int_data['neu'] = my_vader_score_neutral

pd.options.display.max_colwidth = 40


print(int_data)
```
