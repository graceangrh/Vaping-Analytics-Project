```python
govt_ques = pd.read_csv('What do you think the government can do to prevent youths from vaping.csv', header=0, encoding='utf-8')

govt_ques.head(10)
```

```python
govt_ques.rename(columns={'What do you think the government can do to prevent youths from vaping?': 'responses'}, inplace=True)
govt_ques.head(10)
```

```python
govt_ques_longtext = " ".join(text for text in govt_ques.responses)
govt_ques_longtext 
```

```python
stopwords = set(STOPWORDS)
print(stopwords)
```

```python
govt_ques_longtext = govt_ques_longtext.lower()
stopwords.update(["nothing", "haha", "vape", "vaping", "vapes", "make", "will", "singapore", "still", "don", "ppl", "t", "govt"])
print(stopwords)
```

```python
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, max_words=70, background_color="white").generate(govt_ques_longtext)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Save the image in the img folder:
wordcloud.to_file("Govt_Ques.png")
```
