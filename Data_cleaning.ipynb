Cleaning of words

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

#create stop words list
stop_words = nltk.corpus.stopwords.words('english')

#create tokenizer
#\w matches [a-zA-Z0-9_]. This tokenizer splits the string using regular expressions. E.g - '[A-Z]\w+' will select just the capital words
wtk = nltk.tokenize.RegexpTokenizer(r'\w+')

#create lemmatizer
wnl = nltk.stem.wordnet.WordNetLemmatizer()
```

```python
def process_text(sentences):
    process_docs = []
    doc_tokens = []
    for s in sentences:
        s = str(s)
        s = s.lower()
        doc_tokens = wtk.tokenize(s)
        doc_tokens = [token for token in doc_tokens if token not in stop_words]
        doc_tokens = [wnl.lemmatize(token) for token in doc_tokens]
        process_docs.append(" ".join(doc_tokens))
    return process_docs
```

```python
#Q2.12 Qn: Why did you like/not like that your friend interfered?
interfere_reason = list(survey['Q2.12'].values)
cleaned = process_text(interfere_reason)

print(cleaned)
```

```python
survey['cleaned_Q2.12'] = cleaned

survey
```

```python
#Q4.8 Qn: What types of resources do you need to advise your friend from vaping? (e.g. peer helper support or prevention of 
#         vaping related workshops)

interfere_reason = list(survey['Q4.8'].values)
cleaned = process_text(interfere_reason)
survey['cleaned_Q4.8'] = cleaned

survey
```

```python
#Q6.2 Qn: What do you think the government can do to prevent youths from vaping?

govt = list(survey['Q6.2'].values)
cleaned = process_text(govt)
survey['cleaned_Q6.2'] = cleaned

survey
```
