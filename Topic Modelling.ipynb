### Text preprocessing using NLTK

```python
#create tokenizer
#\w matches [a-zA-Z0-9_]. This tokenizer splits the string using regular expressions. E.g - '[A-Z]\w+' will select just the capital words
wtk = nltk.tokenize.RegexpTokenizer(r'\w+')
```

```python
def process_text(documents):
    process_docs = []
    doc_tokens = []
    for doc in documents:
        # doc = doc.lower()
        doc_tokens = wtk.tokenize(doc)
        # doc_tokens = [token for token in doc_tokens if token not in stop_words]
        # doc_tokens = [wnl.lemmatize(token) for token in doc_tokens] # Can try spacy or see how to use the (nltk.pos_tag()) with lemmatizer
        # doc_tokens = [token for token in doc_tokens if len(token) > 1] # Returns words with more than 1 character
        process_docs.append(doc_tokens)
    return process_docs
```

```python
process_reviews = process_text(df2['cleaned_Q6.2'].apply(str))
print(len(process_reviews))
```

```python
# Keep the processed tokens as a sentence
review_sentences = []
for i in range(len(process_reviews)): 
    sentence = process_reviews[i]
    review_sentences.append(' '.join(sentence))

# print(review_sentences)
```

### Topic Modeling

#### 1. Create a dictionary representaion
`gensim` dictionary assigns an integer id to each unique word in the corpus

```python
# Create a dictionary representation of the documents. Each word gets an id
dictionary = gensim.corpora.Dictionary(process_reviews)

print('Sample word to id mappings:\n', list(dictionary.items())[:50])
print('Total Vocabulary Size:', len(dictionary))
```

#### 2.  Set parameters

Filter the vocabulary based on Frequency. We do not want to use less important words or words that do not help differentiate the topics.<br>
https://radimrehurek.com/gensim/corpora/dictionary.html

```python
# Use the reference: Filter out words that occur in less than 20 documents (less impt words),
# or more than 50% of the documents (do not help differentiate topic).

# dictionary.filter_extremes(no_below=20, no_above=0.5)
dictionary.filter_extremes(no_below=2)
print('Total Vocabulary Size:', len(dictionary))
```

#### 3. Create the vector space model (Bag of words model in gensim)
The function `doc2bow()` counts the number of occurrences of each unique word, converts the word to its integer word id and returns the result as a sparse vector (words from the dictionary not present in the document are not stored in the document vector)


```python
# Transforming corpus into bag of words vectors
corpus_vect = [dictionary.doc2bow(text) for text in process_reviews]
```

Check how the vector looks like for first few documents.

```python
# Print a few lines
for i in range(len(corpus_vect)):
    print('Item is ',i)
    print(corpus_vect[i])
    print(process_reviews[i],'\n')
```

#### 4. Apply the GENSIM LDA Model
- https://radimrehurek.com/gensim/models/ldamodel.html. 


```python
# Understanding random numbers
import random

random.seed(1) # always generates same number even across different machines
print(random.randint(0,9))
```

We will be using total of 2 topics.

```python
num_topics = 2
ldamodel = gensim.models.ldamodel.LdaModel(corpus_vect, num_topics = num_topics, id2word=dictionary, passes=25, random_state= 0)
```

```python
print(type(ldamodel))
```

```python
#formatting the topics in print
for num ,topic in ldamodel.show_topics(formatted=True, num_topics=num_topics, num_words=10):
    print(str(num)+": "+ topic)
```

#### 5. Evaluation by Computing the Perplexity and Coherence Score

Perplexity measures the uncertainty, the lower the better. 
Coherence measures "homogeneity", so higher better. 

```python
from gensim.models import CoherenceModel
```

```python
log_perplexity = ldamodel.log_perplexity(corpus_vect)
perplexity = 2**(-log_perplexity)
print('Perplexity: ',perplexity)

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=ldamodel, texts=process_reviews, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('Coherence Score: ', coherence_lda)
```

B) Document as the distribution over words

Interpret the dominant topic(s) for each document. 
Reference: https://www.tutorialspoint.com/gensim/gensim_documents_and_lda_model.htm


```python
def format_topics_sentences(ldamodel=ldamodel, corpus=corpus_vect, texts=review_sentences):
    # Init output
    sent_topics_df = pd.DataFrame()

    # Get main topic in each document
    for i, row_list in enumerate(ldamodel[corpus]):
        row = row_list[0] if ldamodel.per_word_topics else row_list            
        # print(row)
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        # Get the Dominant topic, Perc Contribution and Keywords for each document
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # Add original text to the end of the output
    contents = pd.Series(texts)
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    return(sent_topics_df)
```

Call the defined function.


```python
df_topic_sents_keywords = format_topics_sentences(ldamodel=ldamodel, corpus=corpus_vect, texts=review_sentences)
```

```python
# Format

df_dominant_topic = df_topic_sents_keywords.reset_index()
df_dominant_topic.columns = ['Document_No', 'Dominant_Topic', 'Topic_Perc_Contrib', 'Keywords', 'Text']
df_dominant_topic.head(10)
```
